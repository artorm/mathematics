# 4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1

exp(x) + x∙(1 – y) = 1


```python
from scipy.optimize import fsolve
x = np.linspace(-3, 5, 200)

plt.plot(x,x**2-1)
plt.plot(x,(np.exp(x)-1)/x+1)
plt.ylim(-3,10)
plt.xlim(-3,10)
plt.grid(True)


def question(p):
    x, y = p
    return (y-x**2+1, np.exp(x)+x-x*y-1 )

x1,y1=fsolve(question, (-1,1))
print(x1,y1)

x2,y2=fsolve(question, (2,6))
print(x2,y2)
```
